```python
'''
The following python script is designed for a BJT Miscellaneous Bias Configuration
where Collector feedback is taken while emitter resistance is considered 0.

                Collector Feedback Of BJT

        Fig:4.53
        Page:189
        Book:Electronic Devices and Circuit Theory
            By Robert L. Boylestad, Louis Nashelsky
            11th Edition

        Script Written By:Shirajul Abedin Shimul
        Registration No.2020338001
'''

##We first include our necessary modules for this script
##We are using schemdraw for our assigned task

import schemdraw
import schemdraw.elements as elm      ##Assigning the elements of schemdraw as elm to be called upon

##We assign d as our Drawing canvas for schemdraw elements
d=schemdraw.Drawing()

##Started by creating a BJT npn transistor :
##Made sure that the parameters of our npn BJT is described(ex:amplification factor)
d.config(unit=3)
transistor=elm.BjtNpn(circle=True,fill='None').label(r'$\beta$=120',color='red').fill("#63A7E8")
d+=transistor       ##We added our BJT to the drawing


##Created the 1st resistor or the Collector resistor:
##Made sure that the parameters of our 1st resistor is described

res1=elm.Resistor(at=transistor.collector).up().label('4.7K$\Omega$',loc='bottom',color='red').label('$R_{C}$',loc='top',color='red')
d+=res1            ##We added our resistor to the drawing


##Added a juction point for connection of elements:

dot1=elm.Dot(radius=.02,close=True,at=res1.start)
d+=dot1


##Defined the endpoint of our 1st resistor:
##We labeled endpoint as our Voltage source Vcc=20V connection point

dot2=elm.Dot(radius=.02,close=True,at=res1.end).label('$V_{cc}$=20V',color='red')
d+=dot2


##Created the 2nd resistor or the Base resistor:
##Made sure that the parameters of our 2nd resistor is described

res2=elm.Resistor(at=transistor.collector).left().label('680K$\Omega$',loc='bottom',color='red').label('$R_{B}$',loc='top',color='red')
d+=res2            ##We added our resistor to the drawing


##Created wire for the connections of elements:

l1=elm.Line(at=res2.end).down().length(.7)
d+=l1              ##Added 1st wire to the drawing


l2=elm.Line(at=transistor.base).to(l1.end)    ##Connected transistor base to base resistor through l2 wire
d+=l2              ##Added 2nd wire to the drawing

##Added another junction point:

dot3=elm.Dot(radius=.02,close=True,at=l2.end)
d+=dot3


##Added our 1st capacitor to the BJT base resistor and base junction point:
##We described the parameters for the capacitor

capacitor1=elm.Capacitor2(at=dot3.start).label('10$\mu$F',loc='top',color='red').label('$C_{1}$',loc='bottom',color='red')
d+=capacitor1      ##Added capacitor to the drawing


##Defined the endpoint of our 1st capacitor & its parameters:
##We labeled endpoint as our input Sinusoidal Voltage source (Vi) connection point

dot4=elm.Dot(radius=.02,close=True,at=capacitor1.end).label('$V_{i}$',loc='left',color='red')
d+=dot4


##Defined our 2nd capacitor & its parameters:
##We added it to the 1st junction point or collector terminal of BJT

capacitor2=elm.Capacitor2(at=dot1.start).right().label('10$\mu$F',loc='top',color='red').label('$C_{2}$',loc='bottom',color='red')
d+=capacitor2      ##Added capacitor to the drawing


##Deined endpoint of our 2nd capacitor:
##We labeled endpoint as our output Voltage terminal (Vo) connection point

dot5=elm.Dot(radius=.02,close=True,at=capacitor2.end).label('$V_{0}$',loc='right',color='red')
d+=dot5


##Added Ground to the circuit through emitter of our BJT

Ground=elm.Ground(at=transistor.emitter)
d+=Ground


d.save('Fig4.53_ElecDevice_RBoylstadPg189_11E.pdf') ###Saving Our Project

##Display our drawing

d.draw()
```

```python
#############################################################################################
#############################################################################################
### This python script will produce a DC equivalent circuit of a Direct-coupled Amplifier ###
### Fig.4.72                                                                              ###
### Page: 204                                                                             ###
### Book: Electronic Devices and Circuit Theory                                           ###
### By Robert L. Boylestad, Louis Nashelsky                                               ###
### 11th Edition                                                                          ###
#############################################################################################
### Student_Registration_number: 2020338010                                               ###
#############################################################################################
#############################################################################################



# Using the python package "SchemDraw"
# For more about the SchemDraw package: https://schemdraw.readthedocs.io/en/latest/index.html

# The greek leter is produced with Unicode
# CTRL+Shift+u  then type 03b2 for Beta, 03a9 for Omega, 00b5 for Micro and then press enter key


## Import the necessary packages
import matplotlib.pyplot as plt
import schemdraw
from schemdraw import flow
import schemdraw.elements as e
with schemdraw.Drawing() as ckt:
    #############################################################################################
    # Configuring The circuit
    ckt.config(inches_per_unit=.5, unit=3)

    #############################################################################################
    # adding the first BJT and denoting it by q1

    ckt += (q1 := e.BjtNpn(circle=True, fill='#aaddff',anchor='base').label('$β_1=100$'))

    #############################################################################################
    # Adding the Base line

    ckt += (lb := e.Line().at(q1.base).left().dot())
    ckt += (l1 := e.Line().at(lb.end).up(ckt.unit/3))

    #############################################################################################
    # Adding resistor R1 to the circuit and labeling it

    ckt += (R1 := e.Resistor().up().at(l1.end).label('$R_1$',loc='top').label('33 kΩ',loc='bottom'))

    #############################################################################################
    #adding the DC source

    ckt += e.Line().at(R1.end).up(ckt.unit/3).dot(open='True').label('14 V',loc='right')

    #############################################################################################
    # Adding the Resistor R2 and Ground

    ckt += (l2 := e.Line().at(lb.end).down(ckt.unit/2.6))
    ckt += (r2 := e.Resistor().down().at(l2.end).label('$R_2$',loc='top').label('10 kΩ',loc='bottom'))
    ckt += e.Ground()

    #############################################################################################
    #Adding the Resistor R_E_1 to the emmiter of the first BJT

    ckt += (le := e.Line().at(q1.emitter).down(ckt.unit/6))
    ckt += (re := e.Resistor().down().at(le.end).label('$R_{E_1}$',loc='top').label('2.2 kΩ',loc='bottom'))
    ckt += e.Ground()

    #############################################################################################
    # Adding the Resistor R_c and labeling the point V_B_2

    ckt += (lc := e.Line().at(q1.collector).up(ckt.unit/4).dot().label('$V_{B_2}$'))
    ckt += (rc := e.Resistor().up().at(lc.end).label('$R_c$',loc='top').label('6.8 kΩ',loc='bottom'))

    #############################################################################################
    # Adding the base line to the second BJT

    ckt += (lb2 := e.Line().at(lc.end).right())

    #############################################################################################
    # Adding the Second BJT with labeling

    ckt += (q2 := e.BjtNpn(circle=True, fill='#aaddff',anchor='base').label('$β_2=50$'))

    #############################################################################################
    # Adding the emmiter resistant and ground to the sencond BJT

    ckt += (le2 := e.Line().at(q2.emitter).down(ckt.unit/8).dot().label('$V_{E_2}$',color='#9D44C0'))
    ckt += (re2 := e.Resistor().down().at(le2.end).label('$R_{E_2}$',loc='top').label('1.2 kΩ',loc='bottom'))
    ckt += e.Ground()

    #############################################################################################
    # Connecting the collector terminal of the second BJT to the end terminal of the Collector resistance of the first BJT

    ckt += (lc2 := e.Line().at(q2.collector).up(ckt.unit/9).dot().label('$V_{C_2}$',color='#9D44C0'))
    ckt += (rc2 := e.Line().at(lc2.end).up(ckt.unit/1.5))
    ckt += e.Line().tox(rc.end)
    ckt += (rcc := e.Line().at(rc.end).toy(rc2.end))

    #############################################################################################
    # Adding the DC power source

    ckt += e.Line().at(0.5*(rcc.end+rc2.end)).up(ckt.unit/4).dot(open='True').label('14 V',loc='right')

    #############################################################################################
    # Labeling the Current

    ckt += e.CurrentLabel(length =1, ofst=0.25).at(lb).label("$I_{B_1}$").color('#9D44C0')
    ckt += e.CurrentLabel(length =1, ofst=0.25,reverse='true').at(0.9*(rc.end)).label("$I_{C}$").color('#9D44C0')
    ckt += e.CurrentLabel(length =1, ofst=0.45,reverse='true').at(q1.collector).flip().label("$I_{C_1}$",loc="bottom").color('#9D44C0')
    ckt += e.CurrentLabel(length =0.7, reverse='true',ofst=0.05).at(0.9*(re2.end)).flip().label("$I_{E_2}$").color('#9D44C0')

ckt.save('Fig4.73_ElecDevice_RBoylstadPg204_11E.pdf')
ckt.draw()
plt.show()
```

```python
##################################################################################
##################################################################################
### This python script will produce a common Emitter configuration BJT network ###
### Fig.4.86                                                                ######
### Page: 210                                                               ######
### Book: Electronic Devices and Circuit Theory                             ######
### By Robert L. Boylestad, Louis Nashelsky                                 ######
### 11th Edition                                                            ######
##################################################################################
### Student_Name               : Saeed Ahamed Mridha                        ######
### Student_Registration_number: 2020338012                                 ######
##################################################################################
##################################################################################


# Using the python package "SchemDraw"
# For more about the SchemDraw package: https://schemdraw.readthedocs.io/en/latest/index.html

# The greek leter is produced with Unicode
# CTRL+Shift+u  then type 03b2 for Beta, 03a9 for Omega, 00b5 for Micro and then press enter key



###### Import Library Function #######
import schemdraw
import matplotlib
import schemdraw.elements as elm
with schemdraw.Drawing() as d:
    d.config(unit=3)
    ####### PNP Transistor with label ######
    pnp = elm.BjtPnp(circle=True,fill='cyan').flip().label('$V_{CE}$\t',color='cyan').label(r'$\beta$=120',ofst=(0.9,0.06),fontsize=13).label('+',ofst=(0.01,0.54),color='cyan').label("-",ofst=(0.01,-0.54),color='cyan',fontsize=20)
    d += pnp
    d += elm.Line().length(0.25).at(pnp.emitter).down()
    d += elm.Dot(open=True).label('E',loc='right',fontsize=10)
    d += elm.Line().at(pnp.base).length(0.25).left()
    d += elm.Dot(open=True).label("B",fontsize=10)
    d += elm.Line().length(0.25).at(pnp.collector).up()
    d += elm.Dot(open=True).label('C',loc='right',fontsize=10)

    ######Calloctor to Output terminal Circuit Code (Using Coupling Capacitor)#######
    d +=elm.Line().at(pnp.collector).length(0.60).up()
    d += elm.Dot()
    d.push()
    d += elm.Capacitor2(label='10$\mu$F',fontsize=10).right()
    d += elm.Dot(open=True).label('$v_o$',loc='right',fontsize=10)
    d.pop()

    ###### Collector to Vcc Circuit Code #####
    d += elm.Resistor().up().label(label='2.4k$\Omega$',loc='bot',fontsize=10)
    d += elm.Dot()
    d.push()
    d += elm.Line().length(0.75).right()
    d += elm.Dot(open=True).label('-18V',loc='right',fontsize=10)
    d.pop()
    ###### Vcc to Base Circuit Code #####
    d += elm.Line().left()
    d += elm.Resistor().down().toy(pnp.base).label('47K$\Omega$',loc='bot',fontsize=10)
    d.push()
    d += elm.Line().tox(pnp.base)
    d.pop()
    d.push()

    ###### Input Tarminal to Base Circuit Code (Using Coupling Capacitor)######
    d += elm.Capacitor2(label='10$\mu$F',fontsize=10).left()
    d += elm.Dot(open=True).label('$v_i$',loc='left',fontsize=10)
    d.pop()

    ###### Base to Ground Circut Code #####
    d += elm.Resistor().length(4).down().label('10k$\Omega$',loc='bot',fontsize=10)
    d += elm.Line().length(1.5).right()
    d += elm.Dot()
    d += elm.Ground()

    ###### Ground to Emitter Circuit Code #####
    d += elm.Line().tox(pnp.emitter)
    d += elm.Resistor().label('1.1k$\Omega$',fontsize=10).toy(pnp.emitter)
    d.save('Fig4.86_ElecDevice_RBoylstadPg210_11E.pdf')
    d.draw()
```
