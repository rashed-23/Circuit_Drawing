```python
###############################################################################
###############################################################################
### This python script will produce a JFET voltage divider amplifier network###
### Fig.7.17                                                                ###
### Page: 432                                                               ###
### Book: Electronic Devices and Circuit Theory                             ###
### By Robert L. Boylestad, Louis Nashelsky                                 ###
### 11th Edition                                                            ###
###############################################################################
### Student_Registration_number: 2020338023                                 ###
### Student_Name               : MD. RASHEDUL ISLAM                         ###
###############################################################################
###############################################################################

#Import the necessary packages_______________________________________________________
import schemdraw
import matplotlib
from schemdraw import elements as elm
with schemdraw.Drawing() as d:
    # Basic configuration of drawing environment
    d.config(unit=3)
    #N-type JFet and its terminals___________________________________________________
    jfetn=elm.JFetP(circle=True, fill='skyblue',color='none').reverse()#Backgroud fill JFet
    d.push()
    d+=jfetn #For background fill
    d.pop()
    d += elm.JFetP().reverse()
    d+=elm.Resistor().at(jfetn.drain).up().label('$R_D$', loc="bottom")#Drain resistor
    d.push()#Create anchor at the end of the drain resistor
    d+=elm.Line().tox(jfetn.drain[0] + .5)
    #VDD voltage Source with positive terminal_______________________________________
    d+=elm.Dot(open =True).label('$V_{DD}$',loc="right")
    d.pop()#Back to the anchor point
    d+=elm.Line().left()
    d+=elm.Resistor().down().label('$R_1$',loc="bottom")#First Resistor
    d+=elm.Line().toy(jfetn.gate)
    d.push()#Create a anchor at the end of the first resistor
    d+=elm.Line().tox(jfetn.gate)

    #N-JFet's Terminals ______________________________________________________________
    d += elm.Dot().at((jfetn.gate[0] - .5, jfetn.gate[1])).label('G', ofst=(0, .1))#Gate teminal
    d += elm.Dot().at((jfetn.drain[0], jfetn.drain[1] + .5)).label('D', ofst=(.2, 0))#Drain terminal
    d += elm.Dot().at((jfetn.source[0], jfetn.source[1] - .5)).label('S', ofst=(.2, -.5))#Source terminal
    d.pop()# Back the anchor point
    d.push()# Create another anchor point at the end of the first resistor

    #Input terminal with Capacitor___________________________________________________
    d+=elm.Capacitor2().left().label('$C_1$',loc="bottom")#Input capacitor
    d+=elm.Dot(open=True).label('$V_i$',loc="left",color='skyblue')
    d.pop() #Back to the anchor point
    d+=elm.Dot()
    d+=elm.Resistor().down().label('$R_2$',loc="bottom").toy(-4.5)#Second Resistor
    d+=elm.Ground()#Second Resistor Grounded
    d+=elm.Resistor().at(jfetn.source).down().label('$R_S$', loc="bottom").toy(-4.5)#sourse resistor
    d+=elm.Ground()#Source Resistor Grounded

    # Output terminal with Capacitor_________________________________________________
    d+=elm.Capacitor2().at((jfetn.drain[0], jfetn.drain[1] + .5)).right().label('$C_2$', loc="bottom")#Output capacitor at Drain point
    d+=elm.Dot(open=True).label('$V_O$',loc="right",color='skyblue')
    d+=elm.Line().at((jfetn.source[0], jfetn.source[1] - .5)).right().tox(jfetn.source[0] + 1.75)# Create a line from Source point
    d+=elm.Capacitor2().down().label('$C_S$',loc="bottom").toy(-4.5)#Capacitor parallel with sourse resistor
    d+=elm.Ground()#Capacitor Grounded

    d.save('Fig7.17_ElecDevice_RBoylstadPg432_11E.pdf')
    d.draw()
```
